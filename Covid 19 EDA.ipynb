# Data Analysis

Top 10 Most Affected Location

```python
top_10_infected_locations=df.groupby('location')['total infected count'].sum()
top_10_infected_locations=top_10_infected_locations.sort_values(ascending=False)
top_10_infected_locations[:10]
```

```python
most_affected_locations = top_10_infected_locations[1:11].index
most_affected_locations
```

```python
top_10_infected_locations[:10].plot.barh()
plt.show()
```

Top 10 Death Locations

```python
top_10_death_locations=df.groupby('location')['total death count'].sum()
top_10_death_locations=top_10_death_locations.sort_values(ascending=False)
top_10_death_locations[:10]
```

```python
most_death_locations=top_10_infected_locations[1:11].index
most_death_locations
```

```python
top_10_death_locations.sort_values(ascending = False)[:10].plot.barh()
plt.show()
```

```python
cases=top_10_infected_locations[1:11].values
deaths=top_10_death_locations[most_affected_locations].values
cases_deaths_df=pd.DataFrame({'Location':most_affected_locations, 'Total Cases':cases, 'Total Deaths':deaths})
print(cases_deaths_df)
```

```python
plot_data=pd.melt(cases_deaths_df, id_vars=['Location'], value_vars=['Total Cases', 'Total Deaths'],
                 var_name='Metric', value_name='Case Count')
plt.figure(figsize=(10,7))
sns.barplot(x="Location", hue="Metric", y="Case Count", data=plot_data)
plt.title('Most Affected Locations(top 10)')
plt.show()
```

Comparing the Rate of Increase of the Coronavirus Disease

```python
country1 = 'India'
country1_data = df.loc[df['location'] == country1]
country1_cases = country1_data[['date','total infected count']]

country2 = 'Germany'
country2_data = df.loc[df['location'] == country2]
country2_cases = country2_data[['date','total infected count']]

datewise_cases = country1_cases.merge(country2_cases, how = 'inner', on = 'date') 
print(datewise_cases.tail())
```
