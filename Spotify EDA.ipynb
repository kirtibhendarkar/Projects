# Data Analysis

Top 5 Most Popular Artists

```python
top_5_artists=df.groupby("artists").count().sort_values(by="track_name", ascending=False)["track_name"][:5]
top_5_artists
```

```python
top_5_artists.plot.barh()
plt.show()
```

Top 5 Most Loudest tracks

```python
top_5_loudest_tracks= df[['loudness','track_name']].sort_values(by="loudness", ascending=True)[:5]
top_5_loudest_tracks
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='loudness', y='track_name', data=top_5_loudest_tracks)
plt.title("Top Five Loudest Tracks")
plt.show()
```

Artist with most danceability song

```python
top_5_artists_danceable_songs=df[['danceability', 'track_name', 'artists']].sort_values(by='danceability', ascending=False)[:5]
top_5_artists_danceable_songs
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='danceability', y='artists', data=top_5_artists_danceable_songs)
plt.title("Top 5 Danceable Tracks")
plt.show()
```

Top 10 Instrumentalness tracks

```python
top_10_instrumentalness_tracks=df[["instrumentalness", "track_name", "artists"]].sort_values(by="instrumentalness", ascending=False)[:10]
top_10_instrumentalness_tracks
```

```python
plt.figure(figsize=(12,7))
plt.pie(x="instrumentalness", data=top_10_instrumentalness_tracks, autopct='%1.1f%%', labels=top_10_instrumentalness_tracks.track_name)
plt.show()
```
